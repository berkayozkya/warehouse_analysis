```python
df['Musteri'] = df['Musteri'].str.replace('ü','u')
```

```python
df["durum"] = (df['Miktar'] - df['Sip_Mik'])
df
```

```python
def kontrol(durum):
  if durum <=0 :
    return False 
    #print("yok")  
  else:
    return True
    #print("var")  
```

```python
kontrol = df["durum"].apply(kontrol)
kontrol
```

```python
df["kontrol"] = kontrol
df
```

```python
df["miktar_450"] = (df['Miktar'] // 450)
df["sip_450"] = (df['Sip_Mik'] // 450)
df["kalan_miktar_450"] = (df['Miktar'] % 450)
df["kalan_sip_450"] = (df['Sip_Mik'] % 450)



df
```

```python
df = df[['Musteri', 'Kod', 'Miktar', 'miktar_450', 'kalan_miktar_450', 'Sip_Mik', 'sip_450', 'kalan_sip_450', 'durum', 'kontrol']]
df
```

```python
df["son_kalan"] = (df['miktar_450'] - df['sip_450'] )
df
```

```python
def son_kalann(son_kalan):
  if son_kalan <0 :
    return 0 
  else:
    return son_kalan
   
     
```

```python
son_kalann = df["son_kalan"].apply(son_kalann)
son_kalann
```

```python
df["son_kalan"] = son_kalann
df
```

```python
df.drop(columns=["durum"])
```

```python
df["son_kalan_kg"] = ((df['son_kalan'] *450) + df['kalan_miktar_450'] )
df
```
